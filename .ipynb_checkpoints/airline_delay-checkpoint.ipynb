```python
# Import the required libraries and dependencies

from pathlib import Path
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt

import opendatasets as od
import os
import requests

%matplotlib inline
import hvplot.pandas
```

```python
# download data from Kaggle
od.download("https://www.kaggle.com/datasets/sriharshaeedala/airline-delay")
```

```python
# reading the csv file & dropping the unnecessary columns
file =('airline-delay/Airline_Delay_Cause.csv')
csv_data = (pd.read_csv(file)).drop(columns=['month','carrier','arr_delay', 'carrier_delay', 'weather_delay', 'nas_delay',
       'security_delay', 'late_aircraft_delay'])
```

#### Filter out the Data from 2018 to 2022 to create a Dataframe 'airport_yearly_data' for Performance and Root Cause Analysis

```python
# Slicing yearly data from 2018 to 2022 & dropping null values
airport_yearly_data=csv_data[csv_data["year"].isin([2018,2019,2020,2021,2022])].dropna()
airport_yearly_data.reset_index()

# adding two new columns to the dataframe %_on_time_arrivals & %_delayed_flights
airport_yearly_data["%_on_time_arrivals"]=round(((airport_yearly_data["arr_flights"]-airport_yearly_data["arr_del15"]) / airport_yearly_data["arr_flights"])*100,2)
airport_yearly_data["%_delayed_flight"]=round((airport_yearly_data["arr_del15"]/airport_yearly_data["arr_flights"])*100,2)


# Review the first and last five rows of the DataFrame
display(airport_yearly_data.head())
display(airport_yearly_data.tail())
```

#### Yearly trend in total volume and delayed flights from 2018-2022

```python
# grouping data by year from airport_yearly_data to plot yearly trend of Total arrivals and delays
volume_yearly_group=airport_yearly_data[["year","arr_flights","arr_del15"]].groupby("year").sum()

# using hv line plot to plot to show trend
linechart_yearly=volume_yearly_group.hvplot(xlabel="Year",ylabel="Total Flight/ Flight Delayed",
label="TOTAL ARRIVING FLIGHTS AND FLIGHTS DELAYED BY MORE THAN 15 MINUTES FROM 2018-2022").opts(yformatter="%.0f")

#using hv bar plot to display proportionateof delays to total arrivals
barchart_yearly=volume_yearly_group.hvplot.bar(xlabel="Year",ylabel="Total Flight/ Flight Delayed",
label="TOTAL ARRIVING FLIGHTS AND FLIGHTS DELAYED BY MORE THAN 15 MINUTES FROM 2018-2022",stacked=True).opts(yformatter="%.0f")

# displaying resulted charts
combined_chart= linechart_yearly   +   barchart_yearly
combined_chart
```

#### Building interactive visualization for each airport 


```python
# extracting relevant data from airport dataframe
volume_yearly_group1=airport_yearly_data[["year","airport_name","%_on_time_arrivals","%_delayed_flight"]].groupby(["year","airport_name"]).mean()

# using hv bar plot, group by and stacking columns
yearly_plot_by_airport=volume_yearly_group1.hvplot.bar(groupby="airport_name",x="year",y=["%_on_time_arrivals","%_delayed_flight"],
     xlabel="Year",ylabel="Total Flight/ Flight Delayed",
label="% ON TIME ARRIVALS & DELAYED FLIGHTS BY AIRPORT FROM 2018-2022",stacked=True).opts(yformatter="%.0f",width=700)

#display chart
yearly_plot_by_airport
```

# **Performance Analysis Of 20 Largest Airports from 2018-2022**

```python
# slicing Data Frame for 20 largest airports by total arriving flights
top_20_total_arrival_list = ["ATL","BOS","CLT","DCA","DEN","DFW","DTW","EWR","IAH","JFK","LAS","LAX","LGA","MCO","MSP","ORD","PHX","SEA","SFO","SLC","TPA"]
#creating top 20 DataFrame from the airport data
top_20_total_arrival_df=airport_yearly_data[airport_yearly_data["airport"].isin(["ATL","BOS","CLT","DCA","DEN","DFW","DTW","EWR","IAH","JFK","LAS","LAX","LGA","MCO","MSP","ORD","PHX","SEA","SFO","SLC","TPA"])]
```

```python
# top_20_total_arrival_df_groupby=top_20_total_arrival_df.groupby("airport_name").sum()
# top_20_total_arrival_df_groupby.head(2)
```

```python
# Performance of Flights by top 20 airports

# grouped top 20 air ports by airport_name column
top_20_total_arrival_df_groupby=top_20_total_arrival_df.groupby("airport_name").sum()

# defined labels and values for pie chart
data={'Category':top_20_total_arrival_df_groupby.index,
   'Values': top_20_total_arrival_df_groupby.iloc[:,3]}

df = pd.DataFrame(data)

# Plotting the pie chart
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(df['Values'],autopct='%1.1f%%')

# Hide the autopct labels
for autotext in autotexts:
    autotext.set_visible(True)
    
# Create custom labels for the legend with percentage
labels = ['{0} - {1:1.1f} %'.format(i,j) for i,j in zip(df['Category'], df['Values']/df['Values'].sum()*100)]

# Set the title and legend
ax.set_title('TOP 20 AIR PORT TOTAL ARRIVALS %age')
ax.legend(wedges, labels, title='Categories', loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

### top 20 yearly display

```python
# created a function to display yearly performance of top 20 airports
def top_20_total_yearly_performance_plot(df,year):
    x=df.groupby("airport").sum().sort_values(by="arr_flights",ascending=False
             ).hvplot.bar(y=["arr_flights","arr_del15"
                ],xlabel="AirPort_Name",ylabel="Total Flight/ Flight Delayed"
                     ,label=f"ON TIME ARRIVALS & DELAYED FLIGHTS YEAR {year} FOR TOP 20 AIRPORT ",hover_color="yellow",rot=90
                         ).opts(yformatter="%.0f",width=1000,height=500,multi_level=False)
    return x
```

```python
#created DF
top_20_total_arrival_df_groupby=top_20_total_arrival_df[["year","airport","arr_flights","arr_del15"]].groupby(["year","airport"]).sum()
```

```python

# call function 'top_20_total_yearly_performance_plot' to display 2018 & 2019
top_20_total_yearly_performance_plot(top_20_total_arrival_df_groupby.loc[2018:2019,["arr_flights","arr_del15"]],"2018 & 2019")
```

```python
# call function 'top_20_total_yearly_performance_plot' to display 2020 & 2021
top_20_total_yearly_performance_plot(top_20_total_arrival_df_groupby.loc[2020:2021,["arr_flights","arr_del15"]],"2020 & 2021")
```

```python
# call function 'top_20_total_yearly_performance_plot' to display 2022
top_20_total_yearly_performance_plot(top_20_total_arrival_df_groupby.loc[2022,["arr_flights","arr_del15"]],"2022")
```

# **Root Cause Analysis Of 20 Largest Airports from 2018-2022**

```python
# Numeric Data extracted from top 20 airports data frame to create Pie Chart

# cerated a numeric list
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

# extracted # numeric columns from data frame
pie_chart_df = top_20_total_arrival_df.select_dtypes(include=numerics)

# grouped by the extracted columns by year
pie_chart_df=pie_chart_df.groupby("year").sum()

# dropped unnecessary columns and summed the remaining data by rows 
pie_chart_df=pie_chart_df.drop(columns=["arr_flights","arr_del15", '%_on_time_arrivals', '%_delayed_flight','arr_cancelled', 'arr_diverted']).sum(axis=0)

# created data frame of extracted series
pie_chart_df=pd.DataFrame(pie_chart_df).rename(columns={0:"Values"})


# Pie Plot 

# defined labels and values for pie chart
data={'Category':pie_chart_df.index,
   'Values': pie_chart_df.iloc[:,0]}

df = pd.DataFrame(data)

# Plotting the pie chart
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(df['Values'], labels=None, autopct='%1.1f%%')

# Hide the autopct labels
for autotext in autotexts:
    autotext.set_visible(False)
    
# Create custom labels for the legend with percentage
labels = ['{0} - {1:1.1f} %'.format(i,j) for i,j in zip(df['Category'], df['Values']/df['Values'].sum()*100)]

# Set the title and legend
ax.set_title('Percentage of Flight Delays by Reason')
ax.legend(wedges, labels, title='Categories', loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
